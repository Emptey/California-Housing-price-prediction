### Feature engineering
> In this section I explore the available features to see which ones contribute most to the target variable.
> And through this process, I will be creating new features from existing features.

```python
# creates a new feature for no. of rooms in each hosue
df["RoomsPerHouseHold"] = df["AveRooms"] / df["AveOccup"]
```

```python
# creates a new feature for population per household
df["PopulationPerHouseHold"] = df["Population"] / df["AveOccup"]
```

```python
# create a new feature to identify the region of a property
df["Region"] = np.where(df["Latitude"] > 37, "North", "South")
```

```python
df.sample(3)
```

### Visualiztion
> This section visualizes the data so we can have a better visual representation of the data

```python
sns.scatterplot(x="MedInc", y="MedHouseVal", data=df)
plt.title("Income vs House Price")
plt.show()
```

> The scatterplot above shows the relationship between Income and house prices
> in different neighbourhoods

```python
plt.figure(figsize=(8,5))
sns.regplot(x="MedInc", y="MedHouseVal", data=df, scatter_kws={'alpha':0.3}, line_kws={'color':'red'})
plt.title("Income vs House Price with Trend Line")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.regplot(x="HouseAge", y="MedHouseVal", data=df, scatter_kws={'alpha':0.3}, line_kws={'color':'red'})
plt.title("House Age vs House Price with Trend Line")
plt.show()
```

```python
# Transform Variables with skewed values
df['LogPopulation'] = np.log1p(df['Population'])
df['LogIncome'] = np.log(df['MedInc'])
df
```

```python
df_clean = df[
    (df['MedInc'].between(df['MedInc'].quantile(0.01), df['MedInc'].quantile(0.99))) &
    (df['MedHouseVal'].between(df['MedHouseVal'].quantile(0.01), df['MedHouseVal'].quantile(0.99)))
].copy()
```

```python
df_clean
```

```python
plt.figure(figsize=(8,5))
sns.regplot(x="MedInc", y="MedHouseVal", data=df_clean, scatter_kws={'alpha':0.3}, line_kws={'color':'red'})
plt.title("Income vs House Price with Trend Line")
plt.show()
```

```python
y = df_clean["MedHouseVal"]
```

```python
df_clean
```

```python
# create a new feature that holds binary representation of regions, North=0, South=1 as string data is not scalable.
df_clean["RegionBinary"] = df_clean["Region"].map({"North": 0, "South": 1})
df_clean.sample(10)
```

```python
x = df_clean[['MedInc', 'HouseAge', 'AveRooms', 'AveBedrms',
        'Population', 'AveOccup', 'Latitude', 'Longitude',
        'RoomsPerHouseHold', 
        'PopulationPerHouseHold', 'LogPopulation', 'LogIncome', 
        'RegionBinary']]

```

```python
# Split data into training and testing set, test=20%, train=80%
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
X_train
```

### Scale features
> Within this section, the features are being scaled to reduce bias due to high values

```python
scaler = StandardScaler()
```

```python
X_train_scaled = scaler.fit_transform(X_train)
```

```python
X_test_scaled = scaler.transform(X_test)
```

```python
X_test_scaled
```

```python
rf_model = RandomForestRegressor(
    n_estimators=200,
    random_state=42,
    n_jobs=-1
)
```

```python
rf_model.fit(X_train_scaled,y_train)
```

```python
y_pred = rf_model.predict(X_test_scaled)
```

```python
y_pred
```

```python
print(f"R2 Score: {r2_score(y_test, y_pred)}")
```

```python
print(f"RSME: {mean_squared_error(y_test, y_pred)}")
```

```python
feature_importance = rf_model.feature_importances_
feature_names = X_train.columns
```

```python
features = pd.DataFrame({
    "Features": feature_names,
    "Importance": feature_importance
})
```

```python
features
```

```python
sns.barplot(data=features, x="Importance", y="Features")
```
